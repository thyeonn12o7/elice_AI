```python
# 라이브러리를 호출합니다.
import numpy as np
import pandas as pd
import re
```

```python
# 데이터를 읽고, 데이터프레임 df를 생성합니다.
df = pd.read_csv('./_edit_add_video_address.csv')
df
```

```python
# 문자열을 Datetime으로 변환하기
df['publishedAt'] = pd.to_datetime(df['publishedAt'],format='%Y-%m-%d')
df['trending_date'] = pd.to_datetime(df['trending_date'],format='%Y-%m-%d')
```

```python
# 12/20까지의 데이터만 사용하기
df=df.loc[df['trending_date'] <= '2021-12-20']
```

```python
df
```

```python
# "|" 기호를 콤마로 변경, 리스트로 변환

for i in range(13733):
    df['tags'].iloc[i] = df['tags'].iloc[i].replace('|', ',')
    tag_list = df['tags'].iloc[i].split(',')
    len_tag_list = len(tag_list)

    
    # 리스트를 문자열로 변환
    
    s = ",".join(tag_list)
    df['tags'].iloc[i] = s
```
