```python
def preprocessing(text):
    # ê°œí–‰ë¬¸ì ì œê±°
    text = re.sub('\\\\n', ' ', text)
    return text
```

```python
sentences_title_1 = df1["title"].apply(preprocessing)
sentences_tags_1 = df1["tags"].apply(preprocessing)
```

## 3. ì›Œë“œí´ë¼ìš°ë“œ ë°ì´í„°ì…‹

```python
from soynlp.noun import LRNounExtractor
```

```python
# ë¶ˆìš©ì–´(STOPWORD) ì œê±°
noun_extractor = LRNounExtractor(verbose=True)
noun_extractor.train(sentences_title_1)
nouns_title_1 = noun_extractor.extract()
noun_extractor.train(sentences_tags_1)
nouns_tags_1 = noun_extractor.extract()
```

```python
# List of Tuples
nouns_title_1_list = [(list(nouns_title_1.keys())[0],nouns_title_1[list(nouns_title_1.keys())[0]][0])]

#Create a DataFrame object
nouns_title_1_list = pd.DataFrame(nouns_title_1_list, columns = ['word' , 'value'])

#Add new ROW
for i in range(1,251):
    nouns_title_1_list = nouns_title_1_list.append({'word' : list(nouns_title_1.keys())[i] , 'value' : nouns_title_1[list(nouns_title_1.keys())[i]][0]} , ignore_index=True)

for i in range(0,130):  
    for j in range(0,251):  
        if list(nouns_title_1.keys())[j] == list(nouns_tags_1.keys())[i]:
            nouns_title_1_list['value'] = nouns_title_1_list['value'].replace(nouns_title_1[list(nouns_title_1.keys())[j]][0], 
                                                nouns_title_1[list(nouns_title_1.keys())[j]][0] + nouns_tags_1[list(nouns_tags_1.keys())[i]][0])
        
        else:      
            nouns_title_1_list = nouns_title_1_list.append(pd.DataFrame([[list(nouns_tags_1.keys())[i], nouns_tags_1[list(nouns_tags_1.keys())[i]][0]]], columns=['word', 'value']), ignore_index=True)

        
nouns_title_1_list = nouns_title_1_list.drop_duplicates(['word'])        

nouns_title_1_list = nouns_title_1_list.sort_values(by=['value'], ascending=False)
nouns_title_1_list.to_csv("./wordcloud_set_1.csv",index = False)
nouns_title_1_list
```
