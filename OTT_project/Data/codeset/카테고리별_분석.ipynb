```python
# 라이브러리를 호출합니다.
import numpy as np
import pandas as pd
import re
```

```python
# 데이터를 읽고, 데이터프레임 df를 생성합니다.
df = pd.read_csv('./_edit_add_video_address.csv')
df
```

```python
# 문자열을 Datetime으로 변환하기
df['publishedAt'] = pd.to_datetime(df['publishedAt'],format='%Y-%m-%d')
df['trending_date'] = pd.to_datetime(df['trending_date'],format='%Y-%m-%d')
```

```python
# 12/20까지의 데이터만 사용하기
df=df.loc[df['trending_date'] <= '2021-12-20']
```

```python
df
```

```python
df[df['categoryId']==29]
```

## 1. 기간 나눈 데이터프레임 만들기 - 단계별

```python
# 1단계
df1 = df[df['trending_date'].between('2020-10-11', '2020-11-23') 
         | df['trending_date'].between('2021-11-01', '2021-12-17')]

# 2단계
df2 = df[df['trending_date'].between('2020-08-03', '2020-08-29') 
         | df['trending_date'].between('2020-09-13', '2020-10-10')
         | df['trending_date'].between('2020-11-24', '2020-11-30')
         | df['trending_date'].between('2021-02-15', '2021-07-11')]

# 3단계
df3 = df[df['trending_date'].between('2020-08-30', '2020-09-12') 
         | df['trending_date'].between('2020-12-01', '2021-02-14')]

# 4단계
df4 = df[df['trending_date'].between('2021-07-12', '2021-10-31') 
         | df['trending_date'].between('2021-12-18', '2021-12-20')]
```

## 2. 카테고리별 조회수 좋아요수 댓글수 비교

2.1) 1단계

```python
df1_view_count = pd.DataFrame(df1['view_count'].groupby(df1['categoryId']).sum())
df1_view_count["categoryName"]=['Film & Animation','Autos & Vehicles','Music','Pets & Animals','Sports','Travel & Events','Gaming','People & Blogs','Comedy','Entertainment','News & Politics','Howto & Style','Education','Science & Technology','Society & Issue']
df1_view_count = df1_view_count.sort_index().reset_index()
df1_view_count = df1_view_count[['categoryId', 'categoryName' ,'view_count']]
df1_view_count = df1_view_count.astype({'categoryId': 'int', 'view_count': 'int' ,'categoryName': 'str'})
df1_view_count
```

```python
df1_likes = pd.DataFrame(df1['likes'].groupby(df1['categoryId']).sum())
df1_likes["categoryName"]=['Film & Animation','Autos & Vehicles','Music','Pets & Animals','Sports','Travel & Events','Gaming','People & Blogs','Comedy','Entertainment','News & Politics','Howto & Style','Education','Science & Technology','Society & Issue']
df1_likes = df1_likes.sort_index().reset_index()
df1_likes = df1_likes[['categoryId', 'categoryName' ,'likes']]
df1_likes = df1_likes.astype({'categoryId': 'int', 'likes': 'int' ,'categoryName': 'str'})
df1_likes
```

```python
df1_comment_count = pd.DataFrame(df1['comment_count'].groupby(df1['categoryId']).sum())
df1_comment_count["categoryName"]=['Film & Animation','Autos & Vehicles','Music','Pets & Animals','Sports','Travel & Events','Gaming','People & Blogs','Comedy','Entertainment','News & Politics','Howto & Style','Education','Science & Technology','Society & Issue']
df1_comment_count = df1_comment_count.sort_index().reset_index()
df1_comment_count = df1_comment_count[['categoryId', 'categoryName' ,'comment_count']]
df1_comment_count = df1_comment_count.astype({'categoryId': 'int', 'comment_count': 'int' ,'categoryName': 'str'})
df1_comment_count
```

```python
df1_join = pd.concat([df1_view_count,df1_likes['likes'], df1_comment_count['comment_count']],axis=1)
df1_join.to_csv("./df1_join.csv",index = False)
```
